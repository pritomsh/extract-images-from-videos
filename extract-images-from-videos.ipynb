```python
# Import libraries
import cv2
import os
import matplotlib.pyplot as plt
```

```python
# Read the video from specified path
videoPath = "../input/image-extract/Demo Video.mp4"
vid = cv2.VideoCapture(videoPath)

try:

    # creating a folder named data
    if not os.path.exists('data'):
        os.makedirs('data')

# if not created then raise error
except OSError:
    print('Error: Creating directory of data')
```

```python
# frame
currentframe = 0

while (True):

    # reading from frame
    success, frame = vid.read()

    if success:
        # continue creating images until video remains
        name = './data/frame' + str(currentframe) + '.jpg'
        print('Creating...' + name)

        # writing the extracted images
        cv2.imwrite(name, frame)

        # increasing counter so that it will
        # show how many frames are created
        currentframe += 1
    else:
        break

# Release all space and windows once done
vid.release()
```

```python
def show_images_from_directory(directory_path, num_images=9, file_extension="jpg"):
    # Create a 3x3 grid of subplots
    fig, axs = plt.subplots(3, 3, figsize=(10, 10))
    fig.subplots_adjust(hspace=0.3)  # Adjust vertical space between subplots

    # Use os.listdir() to get a list of file names in the directory
    file_names = os.listdir(directory_path)

    # Filter the file names to get only the ones with the specified extension
    image_names = [name for name in file_names if name.endswith(f".{file_extension}")]

    # Take a subset of images (num_images) in case there are more than 9 images
    image_names = image_names[:num_images]

    for i, image_name in enumerate(image_names):
        image_path = os.path.join(directory_path, image_name)
        # Read the image from the file path using OpenCV
        image = cv2.imread(image_path)

        # Convert the image from BGR to RGB (matplotlib uses RGB format)
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Plot the image on the corresponding subplot
        row = i // 3
        col = i % 3
        axs[row, col].imshow(image_rgb)
        axs[row, col].axis('off')  # Turn off axis labels

        # Use the image name as the title for the subplot
        axs[row, col].set_title(image_name)

    # Display the subplots
    plt.show()

# Call the function to show 9 images from the "/kaggle/working/" directory with the "jpg" extension
directory_path = "/kaggle/working/data/"
show_images_from_directory(directory_path, num_images=9, file_extension="jpg")
```
